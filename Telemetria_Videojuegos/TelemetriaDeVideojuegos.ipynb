```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('videojuegos_online_telemetria.csv')

# Mostrar las primeras filas del DataFrame
print('Primeras 5 filas del DataFrame:')
print(df.head())
```

```python
# Contar el número de valores duplicados en la columna 'PlayerID'
num_duplicates = df['PlayerID'].duplicated().sum()

# Mostrar el resultado
if num_duplicates > 0:
    print(f"Se encontraron {num_duplicates} valores duplicados en la columna 'PlayerID'.")
    print("Las filas duplicadas (manteniendo la primera aparición) son:")
    display(df[df['PlayerID'].duplicated(keep='first')])
else:
    print("No se encontraron valores duplicados en la columna 'PlayerID'.")
```

```python
# Mostrar todas las filas que tienen valores duplicados en la columna 'PlayerID'
# Esto incluye la primera aparición de cada duplicado
duplicated_rows = df[df['PlayerID'].duplicated(keep=False)]

if not duplicated_rows.empty:
    print("Todas las filas con 'PlayerID' duplicados son:")
    display(duplicated_rows)
else:
    print("No se encontraron valores duplicados en la columna 'PlayerID'.")
```

```python
# Eliminar filas duplicadas basándose en la columna 'PlayerID', conservando la primera aparición
df_cleaned = df.drop_duplicates(subset=['PlayerID'], keep='first')

# Mostrar el número de filas antes y después de la eliminación
print(f"Filas antes de eliminar duplicados: {df.shape[0]}")
print(f"Filas después de eliminar duplicados: {df_cleaned.shape[0]}")

# Actualizar el DataFrame original si se desea
df = df_cleaned.copy()

# Verificar si aún existen duplicados en 'PlayerID' para confirmar la eliminación
num_remaining_duplicates = df['PlayerID'].duplicated().sum()
if num_remaining_duplicates == 0:
    print("\nSe han eliminado todos los duplicados en la columna 'PlayerID'.")
else:
    print(f"\nTodavía quedan {num_remaining_duplicates} duplicados en 'PlayerID'.")
```

### Manejo de valores faltantes en la columna 'Age'

Vamos a calcular la mediana de la columna 'Age' y usarla para rellenar los valores faltantes (NaN) en esta columna.

```python
median_age = df['Age'].median()
print(f"Mediana de la columna 'Age': {median_age}")

df.loc[:, 'Age'] = df['Age'].fillna(median_age)

# Verificar si quedan valores NaN en la columna 'Age'
print(f"\nNúmero de valores NaN en la columna 'Age' después de rellenar: {df['Age'].isnull().sum()}")

# Mostrar las primeras filas para verificar el cambio
print("\nPrimeras 5 filas del DataFrame con 'Age' rellenado:")
print(df.head())

# Verificar con df.info() para más detalles
print("\nInformación del DataFrame después de rellenar 'Age':")
df.info()
```

```python
print('Valores de la columna Region antes de los cambios:')
print(df['Region'].value_counts(dropna=False))

# Reemplazar valores NaN por 'USA'
df['Region'] = df['Region'].fillna('USA')

# Reemplazar 'Europe' por 'EU'
df['Region'] = df['Region'].replace('Europe', 'EU')

print('\nValores de la columna Region después de los cambios:')
print(df['Region'].value_counts(dropna=False))

# Verificar las regiones distintas después de los cambios
print('\nRegiones distintas en la columna Region después de los cambios:')
print(df['Region'].unique())
```

```python
distinct_regions = df['Region'].unique()
print(f"Regiones distintas en la columna 'Region': {distinct_regions}")

num_nan_regions = df['Region'].isnull().sum()
print(f"Número de valores NaN (faltantes) en la columna 'Region': {num_nan_regions}")

if num_nan_regions > 0:
    print("Estos valores 'nan' representan datos faltantes que pandas ha detectado o convertido.")

print("\nConteo de valores por cada región (incluyendo NaN):")
print(df['Region'].value_counts(dropna=False))
```

```python
print("Valores distintos en 'FriendsInGame' antes de rellenar NaN:")
print(df['FriendsInGame'].unique())

num_nan_friends_in_game_before = df['FriendsInGame'].isnull().sum()
print(f"\nNúmero de valores NaN (faltantes) en la columna 'FriendsInGame' ANTES de rellenar: {num_nan_friends_in_game_before}")

if num_nan_friends_in_game_before > 0:
    # Calcular la media de FriendsInGame
    mean_friends_in_game = df['FriendsInGame'].mean()
    print(f"Media calculada para 'FriendsInGame': {mean_friends_in_game:.2f}")

    # Rellenar los valores NaN con la media
    df.loc[:, 'FriendsInGame'] = df['FriendsInGame'].fillna(mean_friends_in_game)

    print("\nValores distintos en 'FriendsInGame' DESPUÉS de rellenar NaN:")
    print(df['FriendsInGame'].unique())

    num_nan_friends_in_game_after = df['FriendsInGame'].isnull().sum()
    print(f"\nNúmero de valores NaN (faltantes) en la columna 'FriendsInGame' DESPUÉS de rellenar: {num_nan_friends_in_game_after}")

    if num_nan_friends_in_game_after == 0:
        print("Todos los valores NaN en 'FriendsInGame' han sido rellenados con éxito.")
else:
    print("No se encontraron valores NaN en la columna 'FriendsInGame'.")

print("\nConteo de valores por cada 'FriendsInGame' (incluyendo NaN después de rellenar si aplica):")
print(df['FriendsInGame'].value_counts(dropna=False))
```

```python
print("Valores distintos en 'UsesVoiceChat':")
print(df['UsesVoiceChat'].unique())

num_nan_uses_voice_chat = df['UsesVoiceChat'].isnull().sum()
print(f"\nNúmero de valores NaN (faltantes) en la columna 'UsesVoiceChat': {num_nan_uses_voice_chat}")

if num_nan_uses_voice_chat > 0:
    print("Estos valores 'nan' representan datos faltantes en la columna 'UsesVoiceChat'.")

    # Calcular la proporción de 0s y 1s en los datos existentes
    value_counts = df['UsesVoiceChat'].dropna().value_counts(normalize=True)
    prob_0 = value_counts.get(0, 0) # Probability of 0
    prob_1 = value_counts.get(1, 0) # Probability of 1

    print(f"\nProporción de 0s (no usa chat de voz): {prob_0:.2f}")
    print(f"Proporción de 1s (usa chat de voz): {prob_1:.2f}")

    # Rellenar los valores NaN probabilísticamente
    # Crear una serie de valores aleatorios (0 o 1) basados en las proporciones
    import numpy as np
    fill_values = np.random.choice([0, 1], size=num_nan_uses_voice_chat, p=[prob_0, prob_1])

    # Obtener los índices de los valores NaN
    nan_indices = df[df['UsesVoiceChat'].isnull()].index

    # Asignar los valores rellenados a los NaN
    df.loc[nan_indices, 'UsesVoiceChat'] = fill_values

    print("\nValores distintos en 'UsesVoiceChat' DESPUÉS de rellenar NaN:")
    print(df['UsesVoiceChat'].unique())

    num_nan_uses_voice_chat_after = df['UsesVoiceChat'].isnull().sum()
    print(f"\nNúmero de valores NaN (faltantes) en la columna 'UsesVoiceChat' DESPUÉS de rellenar: {num_nan_uses_voice_chat_after}")

    if num_nan_uses_voice_chat_after == 0:
        print("Todos los valores NaN en 'UsesVoiceChat' han sido rellenados con éxito.")
else:
    print("No se encontraron valores NaN en la columna 'UsesVoiceChat'.")

print("\nConteo de valores por cada 'UsesVoiceChat' (incluyendo NaN después de rellenar si aplica):")
print(df['UsesVoiceChat'].value_counts(dropna=False))
```
